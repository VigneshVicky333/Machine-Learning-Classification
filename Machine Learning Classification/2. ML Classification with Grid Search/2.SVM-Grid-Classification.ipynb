```python
#importing the Libraies
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Reading the Dataset
dataset = pd.read_csv("Social_Network_Ads.csv")
```

```python
dataset
```

```python
dataset=pd.get_dummies(dataset,drop_first=True)
```

```python
dataset
```

```python
indep=dataset[['Age', 'EstimatedSalary', 'Gender_Male']]   
dep=dataset[["Purchased"]]
```

```python
#split into training set and test
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(indep, dep, test_size = 1/3, random_state = 0)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
from sklearn.svm import SVC
```

```python
#https://scikit-learn.org/1.5/modules/generated/sklearn.linear_model.LogisticRegression.html
```

```python
from sklearn.model_selection import GridSearchCV

param_grid = {'kernel':['rbf','poly','sigmoid','linear'],'C':[10,100,1000,2000,3000]}



grid = GridSearchCV(SVC(probability=True), param_grid, refit = True, verbose = 3,n_jobs=-1,scoring='f1_weighted')

# fitting the model for grid search

grid.fit(X_train, y_train)

```

```python
# print best parameter after tuning
#print(grid.best_params_)

re=grid.cv_results_


#print(re)

grid_predictions = grid.predict(X_test)

from sklearn.metrics import confusion_matrix
cm= confusion_matrix(y_test, grid_predictions)

#print classification report

from sklearn.metrics import classification_report
clf_report = classification_report(y_test, grid_predictions)
```

```python
#https://scikit-learn.org/stable/modules/model_evaluation.html --->Using this we can check the evaluation metrics
```

```python
from sklearn.metrics import f1_score
f1_macro=f1_score(y_test, grid_predictions,average='weighted')
print("The f1_macro value for best parameter{}:".format(grid.best_params_),f1_macro)
```

```python
print("The confusion Matrix:\n",cm)
```

```python
print("The report:\n",clf_report)
```

```python
from sklearn.metrics import roc_auc_score

roc_auc_score(y_test,grid.predict_proba(X_test)[:,1])
```
